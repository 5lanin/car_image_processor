# Batch Processing 

## 1. Batch Processing using YOLO BB Detection and SAM

```python
import os
import time
import matplotlib.pyplot as plt
from pathlib import Path
from batch_vehicle_processor import BatchVehicleProcessor


batch_processor = BatchVehicleProcessor(
    segmentation_method='sam2',
    enable_super_resolution=True,
    sr_threshold_size=512,        # Apply SR to regions smaller than 400px
    bbox_expansion=0.25,          # 25% bbox expansion for better context
    region_size=800,              # Standardize all regions to 800x800
    final_size=512,               # Final output size 512x512
    batch_size=16,                 # Process 16 regions simultaneously
    yolo_batch_size=16             # Process 16 images simultaneously with YOLO
)
```

```python
# Get list of test images from the sample directory
image_directory = "/mnt/damian/Projects/car_data_scraper/images/autoevolution_hierarchical/bmw/m3-coupe"
#"/mnt/damian/Projects/car_data_scraper/images/autoevolution_renderings/article_230605"

# Find all image files in the directory
image_paths = []
if os.path.exists(image_directory):
    for filename in os.listdir(image_directory):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_paths.append(os.path.join(image_directory, filename))

print(f"Found {len(image_paths)} images in test directory:")

results = batch_processor.process_image_batch(
    image_paths,
    min_segmentation_score=0.6,
    save_results=True,
    output_dir="./batch_test_output"
)

# plot the results 
all_vehicles = []
for result in results:
    if len(result) > 0:
        for dict_result in result:
            all_vehicles.append(dict_result['processed_image'])

# if len(all_vehicles) > 16:
#     all_vehicles = all_vehicles[:16]    
columns = 8
rows = len(all_vehicles) // columns + (1 if len(all_vehicles) % columns > 0 else 0)

plt.figure(figsize=(15, 15))
for i, img in enumerate(all_vehicles):
    plt.subplot(rows, columns, i + 1)
    plt.imshow(img)
    plt.axis('off')
plt.tight_layout()
plt.show()   
```
