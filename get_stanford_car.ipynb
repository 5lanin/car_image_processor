```python
import os
os.environ["OMP_NUM_THREADS"] = "4"
import PIL
from PIL import Image
import pandas as pd
# from rembg import remove
from datasets import load_dataset

ds = load_dataset("Multimodal-Fatima/StanfordCars_test")
# ds = load_dataset("Multimodal-Fatima/StanfordCars_train")
ds
```

```python
caption = ds['test']['Attributes_ViT_L_14_text_davinci_003_full'][1]
caption
```

```python
goal_dir = "stanford_cars_train"
os.makedirs(f"{goal_dir}/images", exist_ok=True)
car_data_info = {}

for index in range(len(ds['test']['image'])):
    id = ds['test']['id'][index]
    id = int(id) + 8144  # Adjusting ID to match offset by the train set IDs
    filename = f"{id}.png"
    image_path = f"{goal_dir}/images/{filename}"
    if os.path.exists(image_path):
        continue
    image = ds['test']['image'][index]
    if min(image.size) < 100:
        continue
    caption = ds['test']['blip_caption_beam_5'][index]

    image.save(image_path, format="PNG")
    car_data_info[id] = {
        "filename": filename,
        "caption": caption
    }

# save as csv file using pandas
car_data_info_df = pd.DataFrame.from_dict(car_data_info, orient='index')
car_data_info_df.to_csv(f"{goal_dir}/car_data_info.csv", index=False)
```
